## Separando o DataFrame para as tabelas do SQL

```python
lista_pais = list(df['pais'].unique())
lista_jogo = list(df['jogo'].unique())
```

```python
def importacao_lista_sql(list):
    nome_tabela_sql = input('Insira o nome da tabela: ')
    for i in list:
        try:
            query = f"""INSERT INTO {nome_tabela_sql} ({nome_tabela_sql}) VALUES ('{i}')"""
            cursor.execute(query)
            conn.commit()
            print(f'Registro inserido com sucesso: {i}')        
        except Exception as e:
            conn.rollback()
            print(f'Não foi possível inserir o registro: {i}. O erro encontrado foi {e}')
```

```python
importacao_lista_sql(lista_pais)
```

```python
importacao_lista_sql(lista_jogo)
```
